```
import pandas as pd
import numpy as np
```

```
dataframe = pd.read_csv ('TabDesafio.csv')
```

### 1) Apresente as informações sobre quais foram as notas dos alunos em cada questão

```
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

print(dataframe)

# Configurando as duas opções "set_options" para o atributo "pd" para que mostre a tabela inteira, podemos realizar o print do dataframe
```

### 2) Acrescente uma coluna com a nota de cada aluno, que é a soma das notas das questões

```
# Para adicionar uma coluna, com a soma das notas de todos os alunos, usa a seguinte linha de código:

dataframe['Soma_notas'] = dataframe.sum(axis = 1)

# Assim, com o parâmetro "axis = 1" ele irá somar a linha horizontalmente
```

```
dataframe.head()
```

### 3) Apresente o quanto os dados estão dispersos com relação à nota de cada aluno, facilitando, assim, a análise dos dados (considerar o desvio padrão e a variância)

```
# Comando std irá retornar o desvio padrão das notas a partir da coluna "Soma_nota"

dataframe['Soma_notas'].std()
```

```
# Comando "var" irá retornar a variância das notas a partir da coluna "Soma_nota"

dataframe['Soma_notas'].var()
```
